## Classification

```python
def make_sample(X):
    from sklearn.cross_validation import train_test_split

    X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0)

    print("train data shape: %r, train target shape: %r"
      % (X_train.shape, y_train.shape))
    print("test data shape: %r, test target shape: %r"
      % (X_test.shape, y_test.shape))
    return X_train, X_test, y_train, y_test
```

```python
def learn(X_train, y_train, X_test, y_test, C, gamma):
    %%time
    svc = SVC(kernel="rbf", C=C, gamma=gamma, probability=True)
    svc.fit(X_train, y_train)
    predicted = svc.predict(X_test)
    predicted_proba = svc.predict_proba(X_test)
    print classification_report(y_test, predicted)
    return predicted
```

```python
# PCA
X_train, X_test, y_train, y_test = make_sample(X)
predicted_pca = learn(X_train, y_train, X_test, y_test, C=100, gamma=0.01)
```

```python
# PCA
X_train, X_test, y_train, y_test = make_sample(X_pca_scaled)
predicted_pca = learn(X_train, y_train, X_test, y_test, C=100, gamma=0.01)
```

```python
def confusion(predicted, y_test, name):
    plt.imshow(metrics.confusion_matrix(predicted, y_test),
           interpolation='nearest', cmap=plt.cm.binary)
    plt.colorbar()
    plt.xlabel("true label")
    plt.xticks([0,1,2,3], ['1','2','3','4'])
    plt.yticks([0,1,2,3], ['1','2','3','4'])
    plt.ylabel("predicted label");
    plt.savefig('../poster/figures/confusion_{}.pdf'.format(name), transparent=True)
```

## learning curve

```python
train_sizes = np.logspace(2, 3, 5).astype(np.int)
train_sizes
```

```python
n_iter = 6
train_scores = np.zeros((train_sizes.shape[0], n_iter), dtype=np.float)
test_scores = np.zeros((train_sizes.shape[0], n_iter), dtype=np.float)
```

```python
svc = SVC(C=100, gamma=0.1)

for i, train_size in enumerate(train_sizes):
    cv = ShuffleSplit(len(X_iso_scaled), n_iter=n_iter, train_size=train_size)
    for j, (train, test) in enumerate(cv):
        svc.fit(X[train], y[train])
        train_scores[i, j] = svc.score(X[train], y[train])
        test_scores[i, j] = svc.score(X[test], y[test])
```

```python
mean_train = np.mean(train_scores, axis=1)
confidence = sem(train_scores, axis=1) * 2

plt.fill_between(train_sizes, mean_train - confidence, mean_train + confidence,
                color = 'k', alpha = .2)
#plt.plot(train_sizes, mean_train, 'o-k', c='k', label='Train score')

mean_test = np.mean(test_scores, axis=1)
confidence = sem(test_scores, axis=1) * 2

plt.fill_between(train_sizes, mean_test - confidence, mean_test + confidence,
                color = 'b', alpha = .2)
plt.plot(train_sizes, mean_test, 'o-k', c='g', linestyle='--', label='Test score')

plt.xlabel('Training set size')
plt.ylabel('Score')
plt.xlim(0, train_sizes[-1] +100)
plt.ylim((None, 0.99))  # The best possible score is 1.0
#plt.legend(loc='best')
plt.title('Main train and test scores +/- 2 standard errors')
plt.savefig('../poster/figures/learning.pdf', transparent=True)
```
